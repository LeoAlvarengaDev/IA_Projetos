```python
# Bibliotecas Usadas
from sys import displayhook as display
from matplotlib import pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import LabelBinarizer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,cohen_kappa_score , confusion_matrix , roc_auc_score
import numpy as np
```

```python
# Lendo o data frame
df = pd.read_csv('/content/drive/MyDrive/IA-Relatorio/phishing.csv', sep=',')
```

```python
# Realizando a visualização as informações
display(df)
```

```python
# Excluindo a coluna index
df = df.drop(columns=['Index'])
display(df)
```

```python
# Analinisando o tipo de dados presente no data frame
display(df.info())
```

```python
# Trocando os valores que estão  em -1 para 0
df = df.replace(-1, 0)
```

```python
# Seperando as características
x = df.values[:,0:30]
y = df.values[:,30]
```

```python
# Normalizando o array das caracteristicas
scaler = MinMaxScaler()
x = df.values[:,0:30]
x = scaler.fit_transform(x)
print(x)
```

```python
# Realizando a separação teste e treino
# Sendo 80% para treino e 20% para teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x,y,test_size=0.20)
```

```python
# Rede Neural
model = Sequential()
model.add(Dense(10,activation='relu',kernel_initializer='he_normal',input_shape=(30,)))
model.add(Dense(15,activation='relu'))
model.add(Dense(15,activation='relu'))
model.add(Dense(1,activation='sigmoid'))

# compilacao do modelo
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
history = model.fit(x_treino,y_treino,validation_data=(x_teste,y_teste), epochs=50,batch_size=32,verbose=1)
```

```python
# plt.title=('Grafico de custo') # Erros
plt.subplot(211)
plt.plot(history.history['loss'], label='Perdas durante o Treinamento')
plt.plot(history.history['val_loss'], label='Perdas durante os Testes')
plt.xlabel('Epocas')
plt.ylabel('Perdas')
plt.legend()
plt.show()
```

```python
# plt.title('Grafico de Desempenho') # Erros
plt.plot(history.history['val_accuracy'], label='Acurácia durante os Testes')
plt.xlabel('Epocas')
plt.ylabel('Acurácia')
plt.legend()
plt.show()
```

```python
# Enquadrando as respostas gerada pelo modelo
previ = (model.predict(x_teste) > 0.5).astype(np.float32)
display(previ)
```

```python
# Gerando a matrix de confusão
matrix = confusion_matrix(y_teste,previ)
display(matrix)
```

```python
# Plotando a matriz de confusão
plt.imshow(matrix, interpolation='nearest', cmap=plt.cm.Blues)
plt.colorbar()

# Adicionando os valores da matriz
for i in range(matrix.shape[0]):
    for j in range(matrix.shape[1]):
        plt.text(j, i, format(matrix[i, j], 'd'), horizontalalignment="center", color="Black")

plt.xticks(np.arange(2), ['Falso Positivo', 'Falso Negativo'], rotation=0)
plt.yticks(np.arange(2), ['Verdadeiro Positivo', 'Verdadeiro Negativo'])
plt.xlabel('Predito')
plt.ylabel('Real')
plt.show()
```

```python
# accuracy: (tp + tn) / (p + n)
accuracy = accuracy_score(y_teste, previ)
print('Accuracy: %f' % accuracy)
```

```python
# precision tp / (tp + fp)
precision = precision_score(y_teste, previ)
print('Precision: %f' % precision)
```

```python
# recall: tp / (tp + fn)
recall = recall_score(y_teste, previ)
print('Recall: %f' % recall)
```

```python
# f1: 2 tp / (2 tp + fp + fn)
f1 = f1_score(y_teste, previ)
print('F1 score: %f' % f1)
```

```python
# kappa
kappa = cohen_kappa_score(y_teste, previ)
print('Cohens kappa: %f' % kappa)
```

```python
# ROC AUC
auc = roc_auc_score(y_teste, previ)
print('ROC AUC: %f' % auc)
```
