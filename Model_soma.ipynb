```python
# Bibliotecas Usadas
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import LabelBinarizer
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score,cohen_kappa_score , confusion_matrix , roc_auc_score
import numpy as np
import torch
```

```python
# Gerar dados de treinamento e validação
np.random.seed(42)
train_data = [(np.random.randint(0, 10000), np.random.randint(0, 10000)) for _ in range(100000)]

# Preparar os dados em um DataFrame
train_df = pd.DataFrame(train_data, columns=['num1', 'num2'])
train_df['sum'] = train_df['num1'] + train_df['num2']
```

```python
train_df
```

```python
# Seperando as características
x_train_get_feature = train_df.values[:,0:2]
x_train_get_target = train_df.values[:,2]
```

```python
# Realizando a separação teste e treino
# Sendo 80% para treino e 20% para teste
# x_treino, x_teste, y_treino, y_teste = train_test_split(x_train_get_feature_normalizado,x_train_get_target_normalizado,test_size=0.20)
x_treino, x_teste, y_treino, y_teste = train_test_split(x_train_get_feature,x_train_get_target,test_size=0.20)
```

```python
# Rede Neural
model = Sequential()
model.add(Dense(10, activation='relu', kernel_initializer='he_normal', input_shape=(2,)))
model.add(Dense(150, activation='relu'))
model.add(Dense(150, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(150, activation='relu'))
model.add(Dense(15, activation='relu'))
model.add(Dense(1))
```

```python
# Compilação do modelo
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['mean_absolute_error'])
```

```python
# Treinando o modelo
history = model.fit(x_treino,y_treino,validation_data=(x_teste,y_teste), epochs=50,batch_size=32,verbose=1)
```

```python
# Função para realizar a previsão com valores inseridos pelo usuário
def realizar_previsao(model, novos_valores):
    # Certifique-se de que os novos valores tenham o mesmo formato que os dados de treinamento
    novos_valores = np.array(novos_valores).reshape(1, -1)

    # Realizar a previsão
    previsao = model.predict(novos_valores)

    return previsao
```

```python
# Testando a função com valores inseridos pelo usuário
novos_valores = [1000,1520]  # Substitua por seus valores
previsao = realizar_previsao(model, novos_valores)

print("Previsão para os valores inseridos:", previsao)
```

```python
previsao_tensor = torch.tensor(previsao)

# Extraindo o valor como um tipo de dado Python nativo (float)
previsao_valor = int(previsao_tensor.item())

previsao_valor
```
